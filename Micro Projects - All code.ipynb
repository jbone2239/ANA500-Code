```python
df_cleaned = df.dropna(subset=['Year', 'Publisher']).copy()

#convert regional sales into binary classification (1 = Sold, 0 = Not Sold)
df_cleaned.loc[:, 'NA_Sold'] = (df_cleaned['NA_Sales'] > 0).astype(int)
df_cleaned.loc[:, 'EU_Sold'] = (df_cleaned['EU_Sales'] > 0).astype(int)
df_cleaned.loc[:, 'JP_Sold'] = (df_cleaned['JP_Sales'] > 0).astype(int)
df_cleaned.loc[:, 'Other_Sold'] = (df_cleaned['Other_Sales'] > 0).astype(int)

#verify conversion
df_cleaned[['NA_Sales', 'NA_Sold', 'EU_Sales', 'EU_Sold', 'JP_Sales', 'JP_Sold', 'Other_Sales', 'Other_Sold']].head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

#one-hot encode categorical variables
df_encoded = pd.get_dummies(df_cleaned, columns=['Genre', 'Platform', 'Publisher'], drop_first=True)

#select feature columns (excluding raw sales values and non-numeric columns)
features = [col for col in df_encoded.columns if col not in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Name', 'Rank', 'NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']]

#defining target variables for each region
target_columns = ['NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']

#store results
model_results = {}

#loop through each region and train a separate Logistic Regression model
for region in target_columns:
    print(f"\nTraining Logistic Regression model for {region}...\n")
    
    #define X (features) and y (target)
    X = df_encoded[features].copy()
    y = df_encoded[region].copy()

    #standardize numerical features to improve convergence
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    #split into training and testing sets (80% train, 20% test)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    #train Logistic Regression model with increased iterations
    model = LogisticRegression(max_iter=5000, solver='lbfgs')
    model.fit(X_train, y_train)

    #make predictions
    y_pred_proba_logreg = model.predict(X_test)

    #evaluate performance
    accuracy = accuracy_score(y_test, y_pred_proba_logreg)
    report = classification_report(y_test, y_pred_proba_logreg)

    print(f"Accuracy for {region}: {accuracy:.2f}")
    print(f"Classification Report for {region}:\n{report}")

    # Store results in dictionary
    model_results[region] = {'model': model, 'accuracy': accuracy, 'report': report}
```

```python
#dictionary to store feature importance for each region
feature_importance_dict = {}

#loop through each trained model in model_results
for region, result in model_results.items():
    model = result['model']
    feature_coefficients = model.coef_[0]  #get model coefficients
    feature_names = features  #the one-hot encoded features used in training

    #convert to a DataFrame for ease of sorting
    importance_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': feature_coefficients})
    importance_df['Absolute_Coefficient'] = np.abs(importance_df['Coefficient'])  # Take absolute value for ranking
    importance_df = importance_df.sort_values(by='Absolute_Coefficient', ascending=False)

    #store results
    feature_importance_dict[region] = importance_df

    #plot top 10 most important features
    plt.figure(figsize=(10, 5))
    plt.barh(importance_df['Feature'][:10], importance_df['Coefficient'][:10], color='blue')
    plt.xlabel('Coefficient Value')
    plt.ylabel('Feature')
    plt.title(f'Top 10 Most Important Features for {region}')
    plt.gca().invert_yaxis()  #invert axis to show most important first
    plt.grid(True, linestyle="--", linewidth=0.5)
    plt.show()

    print(f"\nTop 10 Features for {region}:\n", importance_df.head(10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
import pandas as pd

#one-hot encode categorical variables
df_encoded = pd.get_dummies(df_cleaned, columns=['Genre', 'Platform', 'Publisher'], drop_first=True)

#select feature columns (excluding raw sales values and non-numeric columns)
features = [col for col in df_encoded.columns if col not in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Name', 'Rank', 'NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']]

#define target variables for each region
target_columns = ['NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']

#dictionary to store SVM results
svm_results = {}

#loop through each region and train an SVM model
for region in target_columns:
    print(f"\nTraining SVM model for {region}...\n")
    
    #define X (features) and y (target)
    X = df_encoded[features].copy()
    y = df_encoded[region].copy()

    #standardize numerical features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    #split into training and testing sets (80% train, 20% test)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    #train SVM model with RBF kernel
    svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
    svm_model.fit(X_train, y_train)

    #make predictions
    y_pred_proba_svm = svm_model.predict(X_test)

    #evaluate performance
    accuracy = accuracy_score(y_test, y_pred_proba_svm)
    report = classification_report(y_test, y_pred_proba_svm)

    print(f"Accuracy for {region}: {accuracy:.2f}")
    print(f"Classification Report for {region}:\n{report}")

    #store results
    svm_results[region] = {'model': svm_model, 'accuracy': accuracy, 'report': report}

```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#ensure categorical variables are one-hot encoded
df_encoded = pd.get_dummies(df_cleaned, columns=['Genre', 'Platform', 'Publisher'], drop_first=True)

#select feature columns (excluding non-numeric columns and raw sales values)
features = [col for col in df_encoded.columns if col not in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 
                                                             'Global_Sales', 'Name', 'Rank', 'NA_Sold', 'EU_Sold', 
                                                             'JP_Sold', 'Other_Sold']]

#standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded[features])

#define target variables for classification
target_columns = ['NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']

#dictionary to store MLP model results
mlp_results = {}

#train an MLP model for each region
for region in target_columns:
    print(f"\nTraining MLP Neural Network for {region}...\n")

    #define target variable
    y = df_encoded[region].values

    #train-test split (80% train, 20% test)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    #MLP Neural Network model
    mlp_model = MLPClassifier(hidden_layer_sizes=(64, 32), activation='relu', solver='adam', max_iter=200, random_state=42, verbose=True)

    #train the model
    mlp_model.fit(X_train, y_train)

    #make predictions
    y_pred = mlp_model.predict(X_test)

    #performance
    accuracy = accuracy_score(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    #display a confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Not Sold', 'Sold'], yticklabels=['Not Sold', 'Sold'])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {region}")
    plt.show()

    #results
    print(f"Accuracy for {region}: {accuracy:.2f}")
    print(f"Classification Report for {region}:\n{class_report}")

    mlp_results[region] = {'model': mlp_model, 'accuracy': accuracy, 'confusion_matrix': cm, 'classification_report': class_report}

print("MLP Neural Network training and evaluation complete.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#categorical variables are one-hot encoded
df_encoded = pd.get_dummies(df_cleaned, columns=['Genre', 'Platform', 'Publisher'], drop_first=True)

#select feature columns (excluding non-numeric columns and raw sales values)
features = [col for col in df_encoded.columns if col not in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 
                                                             'Global_Sales', 'Name', 'Rank', 'NA_Sold', 'EU_Sold', 
                                                             'JP_Sold', 'Other_Sold']]

#standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded[features])

#target variables for classification
target_columns = ['NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']

#convert data to NumPy array and reshape for LSTM input
X_reshaped = np.reshape(X_scaled, (X_scaled.shape[0], 1, X_scaled.shape[1]))

#dictionary to store LSTM model results
lstm_results = {}

#train an LSTM model for each region
for region in target_columns:
    print(f"\nTraining LSTM Model for {region}...\n")

    #define target variable
    y = df_encoded[region].values

    #train-test split (80% train, 20% test)
    X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)

    #define LSTM model
    lstm_model = Sequential([
        LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), activation='relu', return_sequences=True),
        LSTM(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    #compile the model
    lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    #train the model
    lstm_model.fit(X_train, y_train, epochs=10, batch_size=64, validation_split=0.2, verbose=1)

    #make predictions
    y_pred_probs = lstm_model.predict(X_test)
    y_pred = (y_pred_probs > 0.5).astype(int)

    #performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    #display confusion matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Not Sold', 'Sold'], yticklabels=['Not Sold', 'Sold'])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {region}")
    plt.show()

    print(f"Accuracy for {region}: {accuracy:.2f}")
    print(f"Classification Report for {region}:\n{class_report}")

    #store results
    lstm_results[region] = {'model': lstm_model, 'accuracy': accuracy, 'confusion_matrix': cm, 'classification_report': class_report}

print("LSTM Training and Evaluation Complete.")
```

```python
#ROCs
df_encoded = pd.get_dummies(df_cleaned, columns=['Genre', 'Platform', 'Publisher'], drop_first=True)

#feature columns
features = [col for col in df_encoded.columns if col not in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 
                                                           'Global_Sales', 'Name', 'Rank', 'NA_Sold', 'EU_Sold', 
                                                           'JP_Sold', 'Other_Sold']]

#target variables for each region
target_columns = ['NA_Sold', 'EU_Sold', 'JP_Sold', 'Other_Sold']

#standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded[features])

#reshape data for LSTM
X_reshaped = np.reshape(X_scaled, (X_scaled.shape[0], 1, X_scaled.shape[1]))

#dictionary to store ROC curves data
roc_results = {}

#function to plot ROC curves for all models
def plot_roc_curves(region_name):
    plt.figure(figsize=(12, 8))
    plt.title(f'ROC Curves for {region_name}', fontsize=16)
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=14)
    plt.ylabel('True Positive Rate', fontsize=14)
    plt.grid(True)
    
    models_auc = []
    
    #plot ROC for each model
    for model_name, model_data in roc_results[region_name].items():
        plt.plot(model_data['fpr'], model_data['tpr'], 
                 label=f"{model_name} (AUC = {model_data['auc']:.3f})", 
                 lw=2)
        models_auc.append((model_name, model_data['auc']))
    
    #sort models by AUC for the legend
    models_auc.sort(key=lambda x: x[1], reverse=True)
    handles, labels = plt.gca().get_legend_handles_labels()
    order = [labels.index(f"{model[0]} (AUC = {model[1]:.3f})") for model in models_auc]
    plt.legend([handles[i] for i in order], [labels[i] for i in order], 
               loc='lower right', fontsize=12)
    
    plt.tight_layout()
    plt.savefig(f"{region_name}_roc_curves.png", dpi=300, bbox_inches='tight')
    plt.show()
    
    #print models ranked by AUC
    print(f"\nModel Performance Ranking for {region_name} by AUC:")
    for i, (model_name, auc_val) in enumerate(models_auc):
        print(f"{i+1}. {model_name}: AUC = {auc_val:.4f}")

#compute ROC curves for all models across all regions
for region in target_columns:
    print(f"\nComputing ROC curves for {region}...")
    roc_results[region] = {}
    
    #get target variable
    y = df_encoded[region].values
    
    #split data
    X_train_std, X_test_std, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    X_train_lstm, X_test_lstm, _, _ = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)
    
    #for Logistic Regression
    if 'model_results' in globals():
        lr_model = model_results[region]['model']
        y_pred_proba_lr = lr_model.predict_proba(X_test_std)[:, 1]
        fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_proba_lr)
        auc_lr = auc(fpr_lr, tpr_lr)
        roc_results[region]['Logistic Regression'] = {'fpr': fpr_lr, 'tpr': tpr_lr, 'auc': auc_lr}
    
    #for SVM
    if 'svm_results' in globals():
        svm_model = svm_results[region]['model']
        # Use decision_function for SVM if available, otherwise predict
        if hasattr(svm_model, 'decision_function'):
            y_pred_proba_svm = svm_model.decision_function(X_test_std)
        else:
            y_pred_proba_svm = svm_model.predict(X_test_std)
        fpr_svm, tpr_svm, _ = roc_curve(y_test, y_pred_proba_svm)
        auc_svm = auc(fpr_svm, tpr_svm)
        roc_results[region]['SVM'] = {'fpr': fpr_svm, 'tpr': tpr_svm, 'auc': auc_svm}
    
    #for Random Forest
    if 'rf_results' in globals():
        rf_model = rf_results[region]['model']
        y_pred_proba_rf = rf_model.predict_proba(X_test_std)[:, 1]
        fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_proba_rf)
        auc_rf = auc(fpr_rf, tpr_rf)
        roc_results[region]['Random Forest'] = {'fpr': fpr_rf, 'tpr': tpr_rf, 'auc': auc_rf}
    
    #for MLP
    if 'mlp_results' in globals():
        mlp_model = mlp_results[region]['model']
        y_pred_proba_mlp = mlp_model.predict_proba(X_test_std)[:, 1]
        fpr_mlp, tpr_mlp, _ = roc_curve(y_test, y_pred_proba_mlp)
        auc_mlp = auc(fpr_mlp, tpr_mlp)
        roc_results[region]['MLP Neural Network'] = {'fpr': fpr_mlp, 'tpr': tpr_mlp, 'auc': auc_mlp}
    
    #for LSTM
    if 'lstm_results' in globals():
        lstm_model = lstm_results[region]['model']
        y_pred_proba_lstm = lstm_model.predict(X_test_lstm).flatten()
        fpr_lstm, tpr_lstm, _ = roc_curve(y_test, y_pred_proba_lstm)
        auc_lstm = auc(fpr_lstm, tpr_lstm)
        roc_results[region]['LSTM'] = {'fpr': fpr_lstm, 'tpr': tpr_lstm, 'auc': auc_lstm}
    
    #plot ROC curves for this region
    plot_roc_curves(region)

#create combined visualization showing best model for each region
plt.figure(figsize=(14, 10))
plt.title('Best Model ROC Curves by Region', fontsize=18)
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.grid(True)

#for each region, find and plot the best model
for region in target_columns:
    best_model = max(roc_results[region].items(), key=lambda x: x[1]['auc'])
    model_name = best_model[0]
    model_data = best_model[1]
    
    plt.plot(model_data['fpr'], model_data['tpr'], 
             label=f"{region} - {model_name} (AUC = {model_data['auc']:.3f})", 
             lw=3)

plt.legend(loc='lower right', fontsize=12)
plt.tight_layout()
plt.savefig("best_models_by_region_roc.png", dpi=300, bbox_inches='tight')
plt.show()

#summary table of all models' performance by region
summary_data = []
for region in target_columns:
    for model_name, model_data in roc_results[region].items():
        summary_data.append({
            'Region': region,
            'Model': model_name,
            'AUC': model_data['auc'],
        })

summary_df = pd.DataFrame(summary_data)
summary_pivot = summary_df.pivot_table(index='Model', columns='Region', values='AUC')

#ddd"Mean AUC" column
summary_pivot['Mean AUC'] = summary_pivot.mean(axis=1)
summary_pivot = summary_pivot.sort_values('Mean AUC', ascending=False)

#display the summary table
print("\nModel Performance Summary (AUC scores):")
print(summary_pivot.round(4))

#visualization of model performance across regions
plt.figure(figsize=(12, 8))
summary_pivot.drop('Mean AUC', axis=1).T.plot(kind='bar', figsize=(12, 8))
plt.title('Model Performance by Region (AUC)', fontsize=16)
plt.ylabel('AUC Score', fontsize=14)
plt.xlabel('Region', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Model', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.savefig("model_performance_by_region.png", dpi=300, bbox_inches='tight')
plt.show()
```
